```python
def add(a, b):
    return(a+b)
```

```python
def mul(a, b):
    return(a*b)
```

```python
def div(a, b):
    if b == 0:
        return(0.5)
    return(int(a/b))
```

```python
def mod(a, b):
    if a<0 or b<=0:
        return(0.5)
    return(a%b)
```

```python
def eql(a, b):
    return(int(a==b))
```

```python
def change(w, x, y, z, val, which):
    if which == 'w':
        w = val
    elif which == 'x':
        x = val
    elif which == 'y':
        y = val
    else:
        z = val
    return(w, x, y, z)
```
