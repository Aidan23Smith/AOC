```python
def enhanced(enhancer, image):
    decode = {'.':0, '#':1}
    
    new = [[decode[enhancer[0-image[0][0]]] for i in range(len(image[0])+2)] for j in range(len(image)+2)]
    for i in range(1, len(image[0])-1):
        for j in range(1, len(image)-1):
            b = ''
            for k in range(-1, 2):
                for l in range(-1, 2):
                    b += str(image[i+k][j+l])
            new[i+l][j+k] = decode[enhancer[int(b,2)]]
    
    return(new)
```

```python
with open('input1.txt') as f:
    lines = f.read().splitlines()
    
enhancer = lines[0]

image = [[0 for i in range(len(lines[2])+4)] for j in range(len(lines)+2)]

for i in range(2, len(lines)):
    for j in range(len(lines[i])):
        if lines[i][j] == '#':
            image[i][j+2] = 1

for i in range(50):
    image = enhanced(enhancer, image)

sum([x.count(1) for x in image])
```
